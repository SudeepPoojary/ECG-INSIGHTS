```python
import pandas as pd
import numpy as np
import os
from natsort import natsorted
import joblib
```

```python
#creating list to store file_names
NORMAL_=[]
MI_=[]
PMI_=[]
HB_=[]

normal = '/content/drive/MyDrive/MAJOR_PROJECT/preprocessed_data/Normal/csv'
abnormal = '/content/drive/MyDrive/MAJOR_PROJECT/preprocessed_data/Abnormal_hear_beat/csv'
MI = '/content/drive/MyDrive/MAJOR_PROJECT/preprocessed_data/MI/csv'
MI_history = '/content/drive/MyDrive/MAJOR_PROJECT/preprocessed_data/History_MI/csv'

Types_ECG = {'normal':normal,'Abnormal_hear_beat':abnormal,'MI':MI,'History_MI':MI_history}

for types,folder in Types_ECG.items():
  for files in os.listdir(folder):
    if types=='normal':
      NORMAL_.append(files)
    elif types=='Abnormal_hear_beat':
      HB_.append(files)
    elif types=='MI':
      MI_.append(files)
    elif types=='History_MI':
      PMI_.append(files)
```

```python
NORMAL_ = natsorted(NORMAL_)
NORMAL_
```

```python
MI_ = natsorted(MI_)
MI_
```

```python
PMI_ = natsorted(PMI_)
PMI_
```

```python
HB_ = natsorted(HB_)
HB_
```

# **COMBINED CSV OF EACH LEAD(1-12) FROM ALL IMAGES**

```python
#loop over and create combined csv files for each leads.
for x in range(len(MI_)):
  df1=pd.read_csv('/content/drive/MyDrive/MAJOR_PROJECT/preprocessed_data/Normal/csv/{}'.format(NORMAL_[x]))
  df2=pd.read_csv('/content/drive/MyDrive/MAJOR_PROJECT/preprocessed_data/Abnormal_hear_beat/csv/{}'.format(HB_[x]))
  df3=pd.read_csv('/content/drive/MyDrive/MAJOR_PROJECT/preprocessed_data/MI/csv/{}'.format(MI_[x]))
  df4=pd.read_csv('/content/drive/MyDrive/MAJOR_PROJECT/preprocessed_data/History_MI/csv/{}'.format(PMI_[x]))
  final_df = pd.concat([df1,df2,df3,df4],ignore_index=True)
  final_df.to_csv('/content/drive/MyDrive/MAJOR_PROJECT/Combined_csv/Combined_IDLead_{}.csv'.format(x+1))
```

# **NOW COMBINING ALL 12 LEADS INTO A SINGLE CSV FILE**

```python
#now reading just lead1
df=pd.read_csv('/content/drive/MyDrive/MAJOR_PROJECT/Combined_csv/Combined_IDLead_1.csv')
df['Target'].unique()
df.drop(columns=['Unnamed: 0'],inplace=True)
#convert Target column values as Numeric using ngroups
encode_target_label = df.groupby('Target').ngroup().rename("target").to_frame()
test_final  = df.merge(encode_target_label, left_index=True, right_index=True)
test_final.drop(columns=['Target'],inplace=True)
```

```python
#lets try combining all 12 leads in a single csv
location = '/content/drive/MyDrive/MAJOR_PROJECT/Combined_csv'
for files in natsorted(os.listdir(location)):
  if files.endswith(".csv") and not files.endswith("13.csv"):
    if files!='Combined_IDLead_1.csv':
      df=pd.read_csv('/content/drive/MyDrive/MAJOR_PROJECT/Combined_csv/{}'.format(files))
      df.drop(columns=['Unnamed: 0'],inplace=True)
      test_final=pd.concat([test_final,df],axis=1,ignore_index=True)
      test_final.drop(columns=test_final.columns[-1],axis=1,inplace=True)

#drop the target column
test_final.drop(columns=[255],axis=1,inplace=True)
# test_final = test_final.dropna()


# from sklearn.impute import SimpleImputer

# imputer = SimpleImputer(strategy="mean")
# test_final = imputer.fit_transform(test_final)

test_final
```

# **TEST DIMENSIONALITY REDUCTION EXPLAINED VARIANCE ON  THE DATA**

```python
# Now Perform Dimensionality reduction (PCA) on that Dataframe and check
from sklearn.decomposition import PCA


test_final = pd.read_csv('/content/drive/MyDrive/MAJOR_PROJECT/Final_Dataset/Without_Dimentionality_Reduction/final_1D.csv', header=None)



#do PCA and choose componeents as 400
pca = PCA(n_components=400)
x_pca = pca.fit_transform(test_final)
x_pca = pd.DataFrame(x_pca)

# Calculate the variance explained by priciple components
explained_variance = pca.explained_variance_ratio_
print('Variance of each component:', pca.explained_variance_ratio_)
print('\n Total Variance Explained:', round(sum(list(pca.explained_variance_ratio_))*100, 2))

result_df = pd.read_csv('/content/drive/MyDrive/MAJOR_PROJECT/test.csv')

#store the new pca generated dimensions in a dataframe
#store the new pca generated dimensions in a dataframe
pca_df = pd.DataFrame(data = x_pca)
target = pd.Series(result_df.iloc[:,-1], name='target')
final_result_df = pd.concat([pca_df, target], axis=1)
final_result_df
```

```python
#save to dimensionally reduced csv file
final_result_df.to_csv("/content/drive/MyDrive/MAJOR_PROJECT/Final_Dataset/With_Dimentionality_Reduction/pca_final.csv")
```
